### Data Loading

```python
# --------------------------------------------------------------------------------
# 1. Configuration & Tokenizer Initialization
# --------------------------------------------------------------------------------

# [핵심 설정]
# Task 2에서 검증된 최적 길이 512와 6GB VRAM을 위한 배치 사이즈 4
MAX_LEN = 512    
BATCH_SIZE = 4   
MODEL_NAME = 'microsoft/deberta-v3-base'

print(f"Loading {MODEL_NAME} tokenizer...")
# DeBERTa V3는 DebertaV2Tokenizer를 사용해야 합니다.
tokenizer = DebertaV2Tokenizer.from_pretrained(MODEL_NAME)

# --------------------------------------------------------------------------------
# 2. Raw Data Loading
# --------------------------------------------------------------------------------

# CSV 파일 로드
if os.path.exists(TRAINING_SET) and os.path.exists(VALIDATION_SET):
    df_train = pd.read_csv(TRAINING_SET)
    df_validation = pd.read_csv(VALIDATION_SET)
    
    print(f"✅ Raw Data Loaded")
    print(f"   - Training Set: {df_train.shape}")
    print(f"   - Validation Set: {df_validation.shape}")
    print(f"   - Tokenizer initialized: {MODEL_NAME}")
    print(f"   - Max Sequence Length: {MAX_LEN}")
    print(f"   - Batch Size: {BATCH_SIZE}")
else:
    print(f"❌ Error: Dataset files not found at paths:\n{TRAINING_SET}\n{VALIDATION_SET}")
    print("Please check the file paths in the 'Constants' cell.")

# 데이터 미리보기
print("\n--- Train Data Head ---")
print(df_train.head(3))
```

### Tokenization

```python
# --------------------------------------------------------------------------------
# 1. Preprocessing Function Definition
# --------------------------------------------------------------------------------

def preprocess_data(df, tokenizer, max_len):
    """
    DataFrame을 받아서 텍스트를 결합하고, DeBERTa-v3 전용 텐서로 변환합니다.
    - Title과 Plot을 결합하여 문맥 정보를 최대화합니다.
    - 512 토큰 길이로 Truncation(잘라내기)을 수행합니다.
    """
    
    # 1. 텍스트 결합: Title + " " + Plot
    # 데이터프레임의 원본을 건드리지 않기 위해 새로운 리스트로 생성
    texts = df['title'].astype(str) + " " + df['plot_synopsis'].astype(str)
    
    input_ids_list = []
    attention_masks_list = []

    print(f"Tokenizing {len(df)} samples with DeBERTa-v3 tokenizer...")
    
    # 전체 텍스트 토큰화 수행
    for text in texts:
        # encode_plus를 사용하여 Input ID와 Attention Mask를 한 번에 추출
        encoding = tokenizer.encode_plus(
            text,
            add_special_tokens=True,  # [CLS], [SEP] 등 특수 토큰 추가
            max_length=max_len,       # MAX_LEN = 512
            padding='max_length',     # 길이를 512로 맞춤 (Pad 채움)
            truncation=True,          # 512보다 길면 뒤를 자름 (Front Truncation)
            return_attention_mask=True,
            return_tensors='pt',      # PyTorch Tensor로 반환
        )
        
        # 텐서의 차원을 flatten하여 리스트에 추가
        input_ids_list.append(encoding['input_ids'].flatten())
        attention_masks_list.append(encoding['attention_mask'].flatten())

    # 2. 레이블 처리
    # 컬럼 3번(인덱스 3)부터 끝까지가 장르 레이블 (Comedy, Cult, ...)
    # Test 데이터셋(레이블 없음)일 경우를 대비해 예외 처리
    if len(df.columns) > 3:
        labels = torch.tensor(df.iloc[:, 3:].values, dtype=torch.float)
    else:
        labels = None

    # 3. 리스트를 PyTorch 텐서 더미(Stack)로 변환
    input_ids = torch.stack(input_ids_list)
    attention_masks = torch.stack(attention_masks_list)
    
    return input_ids, attention_masks, labels

# --------------------------------------------------------------------------------
# 2. Preprocessing Execution & DataLoader Creation
# --------------------------------------------------------------------------------

# 전처리 실행 (Train / Validation)
# MAX_LEN=512는 Task 2에서 성능 향상이 검증된 설정입니다.
print("--- Processing Training Data ---")
train_input_ids, train_attention_masks, train_labels = preprocess_data(df_train, tokenizer, MAX_LEN)

print("\n--- Processing Validation Data ---")
val_input_ids, val_attention_masks, val_labels = preprocess_data(df_validation, tokenizer, MAX_LEN)

print("\n✅ Preprocessing complete. Tensors created.")

# TensorDataset 생성 (입력값과 정답을 하나로 묶음)
train_dataset = TensorDataset(train_input_ids, train_attention_masks, train_labels)
val_dataset = TensorDataset(val_input_ids, val_attention_masks, val_labels)

# DataLoader 생성
# BATCH_SIZE=4: 6GB VRAM에서 DeBERTa-base를 돌리기 위한 안전한 설정
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=0)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=0)

print(f"✅ DataLoaders Ready.")
print(f"   - Max Sequence Length: {MAX_LEN}")
print(f"   - Batch Size: {BATCH_SIZE}")
print(f"   - Training Batches: {len(train_loader)}")
```

### Model and Training

```python
# --------------------------------------------------------------------------------
# 1. Advanced Training Classes (Innovative Points)
# --------------------------------------------------------------------------------

# [혁신 포인트 1] FGM (Adversarial Training)
# 모델의 임베딩에 노이즈를 주입하여 강건성(Robustness)을 높이는 고급 규제 기법
class FGM:
    def __init__(self, model):
        self.model = model
        self.backup = {}
        
    def attack(self, epsilon=1.0, emb_name='word_embeddings'):
        # 임베딩 가중치를 백업하고 노이즈(Perturbation) 주입
        for name, param in self.model.named_parameters():
            if param.requires_grad and emb_name in name:
                self.backup[name] = param.data.clone()
                norm = torch.norm(param.grad)
                if norm != 0 and not torch.isnan(norm):
                    r_at = epsilon * param.grad / norm
                    param.data.add_(r_at)

    def restore(self, emb_name='word_embeddings'):
        # 가중치 원상 복구
        for name, param in self.model.named_parameters():
            if param.requires_grad and emb_name in name:
                assert name in self.backup
                param.data = self.backup[name]
        self.backup = {}

# [성능 최적화] Focal Loss (Task 2 성공 공식 계승)
class FocalLoss(nn.Module):
    def __init__(self, alpha=0.25, gamma=2.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction
        self.bce = nn.BCEWithLogitsLoss(reduction='none')

    def forward(self, inputs, targets):
        bce_loss = self.bce(inputs, targets)
        pt = torch.exp(-bce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * bce_loss
        return focal_loss.mean() if self.reduction == 'mean' else focal_loss.sum()

# --------------------------------------------------------------------------------
# 2. Model Definition: DeBERTa with Mean Pooling
# --------------------------------------------------------------------------------

class DebertaMeanPoolingClassifier(nn.Module):
    def __init__(self, model_name='microsoft/deberta-v3-base', num_labels=8):
        super(DebertaMeanPoolingClassifier, self).__init__()
        
        # DeBERTa Base 로드
        self.deberta = DebertaV2Model.from_pretrained(model_name)
        self.hidden_size = self.deberta.config.hidden_size # 768
        
        # [혁신 포인트 2] Mean Pooling 적용을 위한 구조 변경
        # [CLS] 토큰 하나만 쓰는 것이 아니라 문장 전체의 평균을 사용
        self.classifier = nn.Linear(self.hidden_size, num_labels)
        self.dropout = nn.Dropout(0.3) # Dropout 0.3 유지

    def mean_pooling(self, token_embeddings, attention_mask):
        # Attention Mask를 사용하여 패딩 토큰은 제외하고 평균 계산
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
        sum_embeddings = torch.sum(token_embeddings * input_mask_expanded, 1)
        sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)
        return sum_embeddings / sum_mask

    def forward(self, input_ids, attention_mask):
        outputs = self.deberta(input_ids=input_ids, attention_mask=attention_mask)
        token_embeddings = outputs.last_hidden_state
        
        # Mean Pooling 적용
        sentence_embedding = self.mean_pooling(token_embeddings, attention_mask)
        
        logits = self.classifier(self.dropout(sentence_embedding))
        return logits

# --------------------------------------------------------------------------------
# 3. Training & Evaluation Functions (FP16 + FGM)
# --------------------------------------------------------------------------------

def train_epoch_task3(model, iterator, optimizer, criterion, device, accumulation_steps, scaler, fgm):
    model.train()
    epoch_loss = 0
    optimizer.zero_grad()

    for i, batch in enumerate(iterator):
        input_ids = batch[0].to(device)
        attention_mask = batch[1].to(device)
        labels = batch[2].to(device)

        # 1. 기본 순전파 (FP16)
        with autocast(device_type='cuda'):
            predictions = model(input_ids, attention_mask)
            loss = criterion(predictions, labels)
            loss = loss / accumulation_steps # Gradient Accumulation

        # 2. 기본 역전파
        scaler.scale(loss).backward()

        # 3. [혁신 포인트] FGM Adversarial Attack (역전파 한 번 더 수행)
        fgm.attack() # 임베딩에 노이즈 추가
        with autocast(device_type='cuda'):
            predictions_adv = model(input_ids, attention_mask)
            loss_adv = criterion(predictions_adv, labels)
            loss_adv = loss_adv / accumulation_steps
        scaler.scale(loss_adv).backward() # 적대적 예제에 대한 Loss 역전파
        fgm.restore() # 임베딩 원상 복구

        # 4. 가중치 업데이트 (Accumulation Steps 마다)
        if (i + 1) % accumulation_steps == 0:
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()
        
        epoch_loss += loss.item() * accumulation_steps

    return epoch_loss / len(iterator)

def evaluate_task3(model, iterator, criterion, device):
    model.eval()
    epoch_loss = 0
    all_probs = []
    all_labels = []

    with torch.no_grad():
        for batch in iterator:
            input_ids = batch[0].to(device)
            attention_mask = batch[1].to(device)
            labels = batch[2].to(device)

            with autocast(device_type='cuda'): # 평가 시에도 FP16 적용
                predictions = model(input_ids, attention_mask)
                loss = criterion(predictions, labels)
            
            epoch_loss += loss.item()
            probs = torch.sigmoid(predictions)
            all_probs.extend(probs.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Dynamic Thresholding Search
    all_probs = np.array(all_probs)
    all_labels = np.array(all_labels)
    best_f1, best_thresh = 0.0, 0.5
    
    for t in np.arange(0.3, 0.71, 0.01):
        preds = (all_probs > t).astype(int)
        score = f1_score(all_labels, preds, average='weighted', zero_division=0)
        if score > best_f1:
            best_f1 = score
            best_thresh = t
            
    return epoch_loss / len(iterator), best_f1, best_thresh

# --------------------------------------------------------------------------------
# 4. Main Execution Block
# --------------------------------------------------------------------------------

# 하이퍼파라미터 (6GB VRAM & DeBERTa Base 최적화)
ACCUMULATION_STEPS = 8   # Batch 4 * 8 = 32 효과
LEARNING_RATE = 1e-5     # 낮은 학습률 필수
N_EPOCHS = 5             # 5 Epoch면 충분

# 모델 초기화
print("Initializing DeBERTa-v3-base with Mean Pooling & FGM...")
model = DebertaMeanPoolingClassifier(num_labels=8).to(device)
print(f'Trainable Parameters: {sum(p.numel() for p in model.parameters() if p.requires_grad):,}')

# 옵티마이저 & 도구 초기화
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=0.0)
criterion = FocalLoss(gamma=2.0, alpha=0.25).to(device)
scaler = GradScaler('cuda')
fgm = FGM(model)

# 학습 루프
best_valid_f1 = 0.0
print(f"\nStarting Task 3 Training (FP16 + FGM + MeanPooling)...")
print("-" * 85)
print(f"{'Epoch':^6} | {'Train Loss':^10} | {'Val Loss':^10} | {'Val F1 (W)':^12} | {'Best Thresh':^12} | {'Msg':^8}")
print("-" * 85)

for epoch in range(N_EPOCHS):
    train_loss = train_epoch_task3(model, train_loader, optimizer, criterion, device, ACCUMULATION_STEPS, scaler, fgm)
    val_loss, val_f1, val_thresh = evaluate_task3(model, val_loader, criterion, device)

    if val_f1 > best_valid_f1:
        best_valid_f1 = val_f1
        torch.save(model.state_dict(), 'best_model_task3.pt')
        saved_msg = "🔥 Saved"
    else:
        saved_msg = ""

    print(f"{epoch+1:02}/{N_EPOCHS:02}  | {train_loss:.4f}     | {val_loss:.4f}     | {val_f1:.4f}       | {val_thresh:.2f}         | {saved_msg}")

print("-" * 85)
print(f"✅ Best Weighted F1: {best_valid_f1:.4f}")
```

### Evaluation scripts

```python
def read_data(submission_file_path, gold_standard_file_path):
    """
    Read submission and gold standard files.
    Extract student ID from filename.
    """
    # Try to find student ID from the filename (looks for 8 digit numbers)
    id_regex = r'\d{8}'

    user_id = re.findall(id_regex, submission_file_path)
    print("Found your ID: ", user_id)
    if user_id:
        user_id = user_id[0]
    else:
        user_id = 'Unknown'

    # Load submission CSV
    print(f"\nLoading submission file: {submission_file_path}")
    submission_df = pd.read_csv(submission_file_path, sep=',', header=None,
                                quoting=csv.QUOTE_NONE, encoding='utf-8')

    # Load gold standard CSV
    print(f"Loading gold standard file: {gold_standard_file_path}")
    gold_standard_df = pd.read_csv(gold_standard_file_path, header=None)

    # Remove columns 1 and 2 (keep only ID and labels)
    gold_standard_df = gold_standard_df.drop([1, 2], axis=1)
    # Skip header row
    gold_standard_df = gold_standard_df.iloc[1:]

    return submission_df, gold_standard_df, user_id


def match_and_prepare_data(submission_df, gold_standard_df, user_id):
    """
    Match submission rows with gold standard rows by ID.
    Prepare data for evaluation.
    """
    gold_standard_labels = []
    submission_labels = []
    missed_rows = []
    submission_df_copy = submission_df.copy()

    print(f"\nMatching submission with gold standard...")
    print(f"Gold standard rows: {len(gold_standard_df)}")
    print(f"Submission rows: {len(submission_df_copy)}")

    # Match each gold standard row with submission
    for index, row in gold_standard_df.iterrows():
        row = row.reset_index(drop=True)
        row_found = False
        row_id = row[0]

        # Extract gold standard labels
        row_labels = [int(row[i]) for i in range(1, len(row))]
        gold_standard_labels.append(row_labels)

        # Find corresponding submission row
        for sub_index, submission_row in submission_df_copy.iterrows():
            if submission_row[0].strip() == row_id.strip():
                try:
                    # Extract submission labels
                    submission_row_labels = [int(submission_row[i]) for i in range(1, len(submission_row))]
                except:
                    # Handle malformed labels (take first character if multi-digit)
                    submission_row_labels = [int(str(submission_row[i])[0]) for i in range(1, len(submission_row))]

                submission_labels.append(submission_row_labels)
                row_found = True
                submission_df_copy.drop(sub_index, inplace=True)
                break

        if not row_found:
            # If row is missing, add inverse labels (worst possible prediction)
            missed_rows.append(row_id)
            submission_labels.append([0 if label == 1 else 1 for label in row_labels])

    return gold_standard_labels, submission_labels, missed_rows


def evaluate_submission(gold_standard_labels, submission_labels):
    """
    Calculate weighted F1 score.
    """
    print(f"\nCalculating weighted F1 score...")

    # Calculate weighted F1 score (accounts for class imbalance)
    f1_weighted = f1_score(gold_standard_labels, submission_labels, average='weighted')

    return f1_weighted


def print_results(user_id, f1_weighted, missed_rows):
    """
    Print evaluation results to screen.
    """
    print("\n" + "="*70)
    print("YOUR SUBMISSION EVALUATION REPORT")
    print("="*70)

    # Alert if ID not found in filename
    if user_id == 'Unknown':
        print('WARNING: ID not found in filename!')
        print('   Please ensure your filename contains your 8-digit student ID.')
        print()

    print(f"Your ID: {user_id}")
    print()

    # Display F1 score with visual indicator
    print("EVALUATION RESULTS:")
    print(f"   Weighted F1 Score: {f1_weighted:.4f}")
    print()

    # Report missing rows
    if missed_rows:
        print(f"MISSING DATA ({len(missed_rows)} rows not found):")
        print("-" * 70)
        for i, row in enumerate(missed_rows[:10], 1):  # Show first 10
            print(f"    {i}. Row ID: {row}")
        if len(missed_rows) > 10:
            print(f"    ... and {len(missed_rows) - 10} more missing rows")
        print()
        print("TIP: Make sure your submission includes all required rows.")
        print("        Missing rows are penalized with worst possible predictions.")
    else:
        print("DATA COMPLETENESS: All expected rows found in your submission!")

    print()
    print("="*70)
    print()


def evaluate(submission_path, gold_standard_path):
    """
    Main function to run the submission evaluation script.
    """

    submission_file = submission_path
    gold_standard_file = gold_standard_path

    # Check if files exist
    if not os.path.exists(submission_file):
        print(f"Error: Your submission file '{submission_file}' not found!")
        print("Make sure the file path is correct and the file exists.")
        sys.exit(1)

    if not os.path.exists(gold_standard_file):
        print(f"Error: Gold standard file '{gold_standard_file}' not found!")
        print("Make sure you have the correct gold standard file.")
        sys.exit(1)

    try:
        # Step 1: Read data
        submission_df, gold_standard_df, user_id = read_data(submission_file, gold_standard_file)

        # Step 2: Match and prepare data
        gold_standard_labels, submission_labels, missed_rows = match_and_prepare_data(
            submission_df, gold_standard_df, user_id
        )

        # Step 3: Evaluate
        f1_weighted = evaluate_submission(gold_standard_labels, submission_labels)

        # Step 4: Print results
        print_results(user_id, f1_weighted, missed_rows)

    except Exception as e:
        print(f"Error during evaluation: {str(e)}")
        print("Please check that your files are in the correct CSV format.")
        print("Each row should contain: ID, label1, label2, label3, ...")
        import traceback
        traceback.print_exc()
        sys.exit(1)
```
